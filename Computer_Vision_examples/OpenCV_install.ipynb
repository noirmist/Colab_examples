## 사전정의된 함수 사용

### BGR to RGB

```
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.grid(None)   
plt.xticks([])
plt.yticks([])
imgplot = plt.imshow(rgb)
```

### BGR to Gray

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.grid(None)   
plt.xticks([])
plt.yticks([])
imgplot = plt.imshow(gray, cmap=plt.cm.gray)
```

## Raw level로 변경

```
height, width, channels = img.shape
print("세로:",height, "가로:", width, "채널:",channels)
```

```
#convert BGR to RGB
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
R_channel = RGB_img.copy()

R_channel[:,:,1] =0
R_channel[:,:,2] =0

plt.grid(None)   
plt.xticks([])
plt.yticks([])
plt.imshow(R_channel)
#plt.show()
```

### 아래의 그림처럼 영상을 만드시오.(각자작성)

<img src=https://i.imgur.com/hDmPZGu.png, height=300px>

```
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Mix_channel = RGB_img.copy()

# 코드 작성
Mix_channel[0:height//2, width//2:, 1] = 0
Mix_channel[0:height//2, width//2:, 2] = 0

Mix_channel[height//2:, width//2:, 0] = 0
Mix_channel[height//2:, width//2:, 2] = 0

Mix_channel[height//2:, :width//2, 0] = 0
Mix_channel[height//2:, :width//2, 1] = 0

plt.grid(None)
plt.xticks([])
plt.yticks([])

plt.imshow(Mix_channel)
plt.show()
```

```
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Mix_channel = RGB_img.copy()

# 코드 작성
for j in range(height):
    for i in range(width):
        if j < height//2 and i > width//2:
            Mix_channel[j,i,1] = 0
            Mix_channel[j,i,2] = 0
        
        if j > height//2 and i > width//2:
            Mix_channel[j,i,0] = 0
            Mix_channel[j,i,2] = 0
        
        if j > height//2 and i < width//2:
            Mix_channel[j,i,0] = 0
            Mix_channel[j,i,1] = 0

plt.grid(None)
plt.xticks([])
plt.yticks([])

plt.imshow(Mix_channel)
plt.show()
```
